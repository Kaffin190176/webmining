### ***Segmentasi Kalimat***

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

def text_segmentation(kalimat):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()
  
  # stemming process
  sentence = kalimat
  output   = stemmer.stem(sentence)
  
  return output

hasil_segmentasi = text_segmentation(process_kalimat)
print(hasil_segmentasi)
```

## ***Case Folding***

```python
# mengubah text menjadi Lower Case

def text_lower(kalimat):
  text = kalimat
  lower_case = text.lower()
  
  return lower_case

hasil_textLower = text_lower(hasil_segmentasi)
print(hasil_textLower)
```

```python
#menghapus angka

import re

def del_number(kalimat):
  text = kalimat
  hasil = re.sub(r"\d+", "", kalimat)
  
  return hasil

hasil_delNumber = del_number(hasil_textLower)
print(hasil_delNumber)
```

```python
#menghapus tanda baca

import string

def del_punctuation(kalimat):
  text = kalimat
  hasil = kalimat.translate(str.maketrans("","",string.punctuation))

  return hasil

hasil_delPunctuation = del_punctuation(hasil_delNumber)
print(hasil_delPunctuation)
```

```python
#menghapus whitespace (karakter kosong)

def del_whitespace(kalimat):
  text = kalimat
  hasil = kalimat.strip()

  return hasil

hasil_caseFolding_delWhitespace = del_whitespace(hasil_delPunctuation)
print(hasil_caseFolding_delWhitespace)
```

## ***Tokenisasi***

```python
from nltk.tokenize import word_tokenize
```

```python
def tokenizing(kalimat):
  text = kalimat
  tokens = nltk.tokenize.word_tokenize(str(kalimat))

  return tokens

hasil_tokenizing = tokenizing(hasil_caseFolding_delWhitespace)
print(hasil_tokenizing)
```

```python
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
#menampilkan informasi frekuensi kemunculan setiap token
def freq_token(kalimat):
  text = kalimat
  tokens = nltk.tokenize.word_tokenize(str(kalimat))
  kemunculan = nltk.FreqDist(tokens)
  
  print(kemunculan.most_common())

  #visualisasi
  import matplotlib.pyplot as plt
  kemunculan.plot(30,cumulative=False)
  plt.show()

freq_token(hasil_tokenizing)
```

## ***Filtering (stopword removal)***

```python
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory
from nltk.tokenize import word_tokenize
import string
```

```python
def filtering(kalimat):
  factory = StopWordRemoverFactory()
  stopword = factory.create_stop_word_remover()
  text = kalimat
  kalimat = kalimat.translate(str.maketrans('','',string.punctuation)).lower()
  stop = stopword.remove(kalimat)
  tokens = nltk.tokenize.word_tokenize(stop)
  
  return tokens

hasil_filtering = filtering(hasil_caseFolding_delWhitespace)
print(hasil_filtering)
```

## ***stemming***

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

```python
def stemming(kalimat):
  factory = StemmerFactory()
  stemmer = factory.create_stemmer()

  text = kalimat 
  hasil = stemmer.stem(str(kalimat))

  return hasil

Hasil = stemming(hasil_filtering)
```

### ***Tf-iDf***

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from sklearn.feature_extraction.text import TfidfVectorizer, TfidfTransformer, CountVectorizer
import pandas as pd
```

```python
def dummy(kalimat):
  kalimat = process_kalimat
  
  return kalimat

tfIdfVectorizer=TfidfVectorizer(tokenizer=dummy,use_idf=True,lowercase=False)
tfIdf = tfIdfVectorizer.fit_transform(tokenizing(Hasil))
df = pd.DataFrame(tfIdf[0].T.todense(), index=tfIdfVectorizer.get_feature_names_out(), columns=["TF-IDF"])
df = df.sort_values('TF-IDF', ascending=False)
print (df)
```
